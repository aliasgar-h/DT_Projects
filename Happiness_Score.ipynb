Above visualizations shows small amount of outliers in Family,Generosity and Dystopis Residual column while in Trust lots of outliers are present which needs to be removed

# Outliers Removal

```python
#Importing Zscore from Scipy
from scipy.stats import zscore
```

```python
#Evaluating ZScore
z=np.abs(zscore(hapscore_col))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
hapscore_col_new=hapscore_col[(z<3).all(axis=1)]
```

As the skewness is removed now we can move ahead for training and test split process

# Finding best random state

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
#Defining Imput and Target data to x & y rspectively
x=hapscore_col_new.drop('Happiness Score',axis=1)
y=hapscore_col_new.iloc[:,-1]
```

```python
x
```

```python
x.shape
```

```python
y
```

```python
y.shape
```

```python
maxAccu=0
maxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=i)
    LR=LinearRegression()
    LR.fit(x_train,y_train)
    LRpred=LR.predict(x_test)
    acc=r2_score(y_test,LRpred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print('Best R2 Score is : ', maxAccu, ' when Random state is : ',maxRs)
```

# Creating Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=37)
```

# Importing Regression Models for Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
```

```python
#Importing error and performance metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
#Using LinearRegression model for training
LR=LinearRegression()
LR.fit(x_train,y_train)
LR_pred=LR.predict(x_test)

#Identifying Errors and R2 Score to check the error in values and the variance in predicted value(Y) with respect to chang in input values(X)

print('Squared mean error in the prediction is : ',mean_squared_error(y_test,LR_pred))
print('Absolute mean error in the prediction is : ',mean_absolute_error(y_test,LR_pred))
print('Root Squared mean error in the prediction is : ',np.sqrt(mean_squared_error(y_test,LR_pred)))
```

```python
#Displayig R2 Score of the model
print('R2 Score of the model is : ',r2_score(y_test,LR_pred))
```

```python
#Using Support Vector Regressor
svr=SVR()
svr.fit(x_train,y_train)
svr_pred=svr.predict(x_test)

#Identifying Errors and R2 Score to check the error in values and the variance in predicted value(Y) with respect to chang in input values(X)

print('Squared mean error in the prediction is : ',mean_squared_error(y_test,svr_pred))
print('Absolute mean error in the prediction is : ',mean_absolute_error(y_test,svr_pred))
print('Root Squared mean error in the prediction is : ',np.sqrt(mean_squared_error(y_test,svr_pred)))
```

```python
#Displayig R2 Score of the model
print('R2 Score of the model is : ',r2_score(y_test,svr_pred))
```

```python
#Using Decision Tree Regressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtr_pred=dtr.predict(x_test)

#Identifying Errors and R2 Score to check the error in values and the variance in predicted value(Y) with respect to chang in input values(X)

print('Squared mean error in the prediction is : ',mean_squared_error(y_test,dtr_pred))
print('Absolute mean error in the prediction is : ',mean_absolute_error(y_test,dtr_pred))
print('Root Squared mean error in the prediction is : ',np.sqrt(mean_squared_error(y_test,dtr_pred)))
```

```python
#Displayig R2 Score of the model
print('R2 Score of the model is : ',r2_score(y_test,dtr_pred))
```

We are getting the best score with Linear Regression model (99%). We check the models with cross validation for any overfitting

```python
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(LR,x,y,cv=5)
print('Cross Validation Score of Linear Regression model is : ',cvs.mean())
```

```python
cvs=cross_val_score(svr,x,y,cv=5)
print('Cross Validation Score of Linear Regression model is : ',cvs.mean())
```

```python
cvs=cross_val_score(dtr,x,y,cv=5)
print('Cross Validation Score of Linear Regression model is : ',cvs.mean())
```

On the basis of R2 score and Cross validation score we can say Linear Regression model has the best result.
